```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
```

```python
iris_data.keys()
```

```python
X = iris_data['data']
y = iris_data['target']
```

```python
print(X.shape)
print(y.shape)
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(solver='newton-cg').fit(X, y)
```

```python
clf.score(X, y)
```

# Задача First-break picking

Анализируется распространение волны в твёрдых телах или жидкостях.
Делают взрыв (shot), затем геофонами, выстроенными в линию (приёмни-
ки, recievers) принимают отражённый сигнал.

Каждый геофон записывает сигнал во времени (trace). Таким образом,
для каждой линии геофонов получаем матрицу, которую можно интерпре-
тировать как 2D-изображение (по горизонтали расположение, по вертикали
время).
Но таких линий геофонов может быть несколько на разном расстоянии
(обычно параллельные линии) и они все принимают сигнал от одного и
того же источника
